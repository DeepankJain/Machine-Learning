```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
df['sex'] = np.where(df['sex'] == 'male', 0, 1)
df.head()
```

```python
df['embarked'] = np.where(df['embarked'] == 'S', 0, np.where(df['embarked'] == 'C', 1, np.where(df['embarked'] == 'Q', 2, 0)))
df.head()
```

```python
#cleaning dataset
df = df[['survived', 'pclass', 'sex', 'age', 'parch', 'embarked']].dropna(axis = 0, how = 'any')
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split as tts
train, test= tts(df, test_size = 0.3)
```

```python
xtrain = train.loc[:, train.columns != 'survived']
xtest = test.loc[:, test.columns != 'survived']
ytrain = train.loc[:, train.columns == 'survived']
ytest = test.loc[:, test.columns == 'survived']
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(xtrain, ytrain)
```

```python
ypred = gnb.predict(xtest)
```

```python
result = pd.DataFrame({'Actual': ytest, 'Predicted': ypred})
print(result)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(ytest, ypred)
```
