```
#importing libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from collections import Counter
```

```
#loading dataset
iris = sns.load_dataset('iris')
iris.head()
```

```
#Univariate data plotting
#histogram
sns.distplot(iris['sepal_length'])
plt.show()
```

```
plt.close()
```

```
sns.distplot(iris['sepal_width'])
plt.show()
```

```
plt.close()
```

```
plt.subplot(221)
sns.distplot(iris['sepal_length'])
plt.subplot(222)
sns.distplot(iris['sepal_width'])
plt.subplot(223)
sns.distplot(iris['petal_length'])
plt.subplot(224)
sns.distplot(iris['petal_width'])
plt.show()
```

```
plt.subplot(211)
sns.kdeplot(iris['sepal_length'])
plt.subplot(212)
sns.kdeplot(iris['sepal_width'])
plt.show
```

```
plt.close()
```

```
plt.close()
```

```
#plotting bivariate plots
#scatter plot of iris dataset
sns.jointplot(data = iris, x = 'sepal_length', y = 'petal_length')
plt.show()
```

```
#kernel density estimation for bivariate data
sns.jointplot(x = iris['petal_length'], y = iris['petal_width'], kind = 'hex', color = 'black')
plt.show()
```

```
sns.jointplot(x = iris['petal_length'], y = iris['petal_width'], kind = 'kde', color = 'blue')
plt.show()
```

```
plt.close()
```

```
sns.pairplot(iris)
plt.show()
```

```
plt.close()
```

```
#strip plot 
sns.stripplot(x = 'species',y = 'sepal_length', data = iris)
plt.show()
```

```
plt.close()
```

```
sns.swarmplot(x = 'species' , y = 'sepal_length', data = iris)
plt.show()
```

```
plt.close()
```
